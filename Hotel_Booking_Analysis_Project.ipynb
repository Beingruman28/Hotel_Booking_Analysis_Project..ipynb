#### 1. Booking Patterns: Analyzing the distribution of bookings across different months, weekdays/weekends to identify seasonal or temporal patterns..

```python
# Booking Patterns: Distribution of Bookings by Month
plt.figure(figsize=(8, 6))
sns.countplot(data=hotel_booking_analysis_df, x='arrival_date_month')
plt.xlabel('Month')
plt.ylabel('Booking Count')
plt.title('Distribution of Bookings by Month')
plt.xticks(rotation=45)
plt.show()
```

```python
# Booking Patterns: Distribution of Bookings by Weekday/Weekend
# Calculate the total number of bookings for weekdays and weekends for each hotel type
hotel_booking_analysis_df['total_nights'] = hotel_booking_analysis_df['stays_in_weekend_nights'] + hotel_booking_analysis_df['stays_in_week_nights']

# For city hotel
city_weekday_bookings = hotel_booking_analysis_df[(hotel_booking_analysis_df['hotel'] == 'City Hotel') & (hotel_booking_analysis_df['stays_in_week_nights'] > 0)]['total_nights'].count()
city_weekend_bookings = hotel_booking_analysis_df[(hotel_booking_analysis_df['hotel'] == 'City Hotel') & (hotel_booking_analysis_df['stays_in_weekend_nights'] > 0)]['total_nights'].count()

# For resort hotel
resort_weekday_bookings = hotel_booking_analysis_df[(hotel_booking_analysis_df['hotel'] == 'Resort Hotel') & (hotel_booking_analysis_df['stays_in_week_nights'] > 0)]['total_nights'].count()
resort_weekend_bookings = hotel_booking_analysis_df[(hotel_booking_analysis_df['hotel'] == 'Resort Hotel') & (hotel_booking_analysis_df['stays_in_weekend_nights'] > 0)]['total_nights'].count()

# Visualize the distribution of bookings for both hotel types
booking_counts = [[city_weekday_bookings, resort_weekday_bookings], [city_weekend_bookings, resort_weekend_bookings]]
booking_labels = ['Weekdays', 'Weekends']
hotel_labels = ['City Hotel', 'Resort Hotel']

fig, ax = plt.subplots()
bar_width = 0.35
opacity = 0.8

index = range(len(booking_labels))

# Plotting the bar chart
rects1 = plt.bar(index, booking_counts[0], bar_width, alpha=opacity, color='b', label=hotel_labels[0])
rects2 = plt.bar([i + bar_width for i in index], booking_counts[1], bar_width, alpha=opacity, color='g', label=hotel_labels[1])

plt.xlabel('Booking Type')
plt.ylabel('Number of Bookings')
plt.title('Bookings: Weekdays vs Weekends - Hotel Types')
plt.xticks([i + bar_width/2 for i in index], booking_labels)
plt.legend()

plt.tight_layout()
plt.show()
```

These visualizations provide insights into the booking patterns:

 * Distribution of Bookings by Month: This bar plot shows the number of bookings made in each month, allowing us to identify any seasonal or monthly booking trends.

 * Distribution of Bookings by Weekday/Weekend: This bar plot shows the count of bookings based on the number of weekend nights stayed, differentiated by hotel type. It helps us understand the preference for weekend stays and the distribution between resort hotels and city hotels.
